```python
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
import math
```

```python
img = np.array(Image.open(".\\Dog.jpg"))


def NN_Interpolation_Resize(img,outWidth,outHeight):
    
    # Get Input Image Width and Height 
    inHeight = img.shape[0]
    inWidth = img.shape[1]
    Input_dims = img.shape[2]
    
    
    # Find the Ratio between input and output image
    rowRatio = inWidth/outWidth
    colRatio =inHeight/outHeight
    
    
    # Create an empty image of output size
    Output_Image = np.zeros((outHeight,outWidth,Input_dims),dtype=np.uint8)


    # For evry dimension (colorts RGB ) in an image 
    for dim in range(Input_dims):
        for col in range(outHeight):
            for row in range(outWidth):
                srcx = round(col*colRatio)
                srcy = round(row*rowRatio)
                try:
                    Output_Image[col,row,dim] =img[srcx,srcy,dim]
                except:
                    continue
    return Output_Image

pil_img = NN_Interpolation_Resize(img,800,800)    
    
plt.imshow(pil_img)
```
